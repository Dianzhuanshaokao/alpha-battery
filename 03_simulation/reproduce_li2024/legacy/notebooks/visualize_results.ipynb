# Li2024 仿真结果可视化\n\n此 notebook 用于读取 `results/<case>/Excel/*.xlsx` 并绘制关键老化指标。

```python
from pathlib import Path\nimport pandas as pd\nimport matplotlib.pyplot as plt\n\nplt.rcParams['figure.figsize'] = (7, 4)\nplt.rcParams['figure.dpi'] = 120
```

```python
# 修改为你希望分析的 case 目录\nCASE_DIR = Path('../results/okane2023_exp3_25C').resolve()\nexcel_dir = CASE_DIR / 'Excel'\nxlsx_files = sorted(excel_dir.glob('*.xlsx'))\nif not xlsx_files:\n    raise FileNotFoundError(f'No xlsx files found in {excel_dir}')\n\ndf_all = pd.concat([pd.read_excel(f) for f in xlsx_files], ignore_index=True)\ndf_all.head()
```

```python
candidate_x = ['Throughput capacity [kA.h]', 'Scan No']\ncandidate_y = ['CDend SOH [%]', 'CDend LLI [%]', 'CDend LAM_ne_tot [%]', 'CDend LAM_pe_tot [%]']\n\nx_col = next((c for c in candidate_x if c in df_all.columns), None)\nif x_col is None:\n    raise KeyError(f'Cannot find x axis column in {candidate_x}')\n\nfor y_col in candidate_y:\n    if y_col in df_all.columns:\n        ax = df_all.plot(x=x_col, y=y_col, marker='o', legend=False)\n        ax.set_title(f'{y_col} vs {x_col}')\n        ax.grid(True, alpha=0.3)\n        plt.show()
```
